## Handling Outlier atau Mengatasi Outlier menggunakan metode IQR

```python
Q1 = df['home_team_fifa_rank'].quantile(0.25)
Q3 = df['home_team_fifa_rank'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['home_team_fifa_rank'] >= lower_bound) & (df['home_team_fifa_rank'] <= upper_bound)]

plt.figure(figsize=(8,6))
sns.boxplot(df_no_outliers['home_team_fifa_rank'])

plt.title('Boxplot setelah Menghapus Outliers')
plt.xlabel('Kolom')
plt.ylabel('Rank')

plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data agar nilai piksel berada di rentang [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
```

```python
# Mengubah label menjadi one-hot encoding
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

## Mengompilasi Model

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```
